# Filtering with .where() method

- Return original dataframe

- Those rows which have the matching condition will show all the values & if not match than it will be filled with <b>NaN</b>.

```python
mask = df['Actor'] == "Sean Connery"

df[mask]
```

```python
df.where(mask)
```

```python
# Show "Box Office Gross > 800"

df.where(df['Box Office'] > 800)
```

```python
mask = df['Actor'] == "Sean Connery"
mask2 = df['Box Office'] > 800

df.where(mask & mask2)
```

```python
import pandas as pd

df = pd.read_csv(r"D:\courses\Udemy - Data Analysis with Pandas and Python data\All the Files\jamesbond.csv",
                 index_col='Film')

df.sort_index(inplace=True)

df.head()
```

```python
columns = ['Box Office', 'Budget', 'Bond Actor Salary']

for i in columns:
    df[i] = df[i].apply(convert_string_to_million)
    
df.head()
```

```python
def good_move(row):
    
    Actor = row[1]
    Budget = row[4]
    
    if Actor == "Pierce Brosnan":
        return "The best"
    
    elif Actor == "Roger Moore" and Budget > '40.0':
        return "Enjoyable"
    
    else:
        return "Not so good"
    
df.apply(good_move, axis=1)
```
